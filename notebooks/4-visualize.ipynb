## Visualize Epsilon

```python
epsilons = []
min_epsilon = 0.1
max_epsilon = 0.9
max_episodes = 160
for i in range(max_episodes):
    epsilon = max(min_epsilon, max_epsilon - (max_epsilon - min_epsilon) * (i / (0.6 * max_episodes)))
    epsilons.append(epsilon)

from matplotlib import pyplot as plt
plt.plot(epsilons)
plt.title('Epsilon Decay')
plt.xlabel('Episodes')
plt.ylabel('Epsilon')
plt.show()
```
